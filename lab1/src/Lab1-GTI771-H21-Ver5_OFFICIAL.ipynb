```python
from skimage.feature import local_binary_pattern
from skimage import filters
from sklearn.decomposition import PCA
```

```python
fer2013 = pd.read_csv('fer2013.csv')
fer2013.head()

df = fer2013
df['lbp_pixels'] = 0
df['sobel_pixels'] = 0
df['lbp_pca'] = 0
df['sobel_pca'] = 0
```

```python
# Convert string array to int array
def string_toarray(x):
    string_array = np.array(x['pixels'].split(' '))
    x['pixels'] = np.array([int(s) for s in string_array])
    return x

df = df.apply(string_toarray, axis =1)
```

### LBP Feature extraction 

```python
# Extract LBP features from pixels
radius = 1
n_points = 8 * radius  
def lbp(x):
    x['lbp_pixels'] = local_binary_pattern(x['pixels'].reshape(48,48), n_points, radius).reshape(-1)
    return x

df = df.apply(lbp, axis =1)
```

### SOBEL Feature extraction 

```python
# Extract SOBEL features from pixels
def sobel(x):
    x['sobel_pixels'] = filters.sobel(x['pixels'].reshape(48,48)).reshape(-1)
    return x

df = df.apply(sobel, axis =1)
```

### PCA reduction of feature N=2

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)

# PCA
lbp_pca = pca.fit_transform(df['lbp_pixels'].tolist())
df['lbp_pca'] = lbp_pca.tolist()

sobel_pca = pca.fit_transform(df['sobel_pixels'].tolist())
df['sobel_pca'] = sobel_pca.tolist()

df.head(2)
```

```python
lbp = df[['emotion','lbp_pixels','Usage']]
sobel = df[['emotion','sobel_pixels','Usage']]

lbp.to_csv('fer2013_lbp.csv',index=False)
sobel.to_csv('fer2013_sobel.csv',index=False)
```

```python
import matplotlib.pyplot as plt
def getLabel(id):
    return ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral'][id]
plt.figure(figsize=(10, 10))
j = 0
for i in [5, 40, 70, 105, 135, 171, 191]:
    plt.subplot(330 + 1 + j)
    plt.imshow(np.squeeze(X_test_jaffe[i], axis=0),cmap=plt.get_cmap('gray'))
    plt.gca().get_xaxis().set_ticks([])
    plt.gca().get_yaxis().set_ticks([])
    plt.ylabel('Label = %s' % getLabel(y_test_jaffe[i]), fontsize = 12)
    j = j + 1
plt.show()
```
